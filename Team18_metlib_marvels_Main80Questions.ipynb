## 76 What is the ratio of those with congestive heart failure to those who had type of heart failure Both?



```python

# Calculate the count of rows where congestive heart failure is indicated
count_congestive_heart_failure = (df_CardiacComplication['congestive_heart_failure'] == 1).sum()

# Calculate the count of rows where the type of heart failure is 'Both'
count_heart_failure_Both = (df_CardiacComplication['type_of_heart_failure'] == 'Both').sum()

# Calculate the ratio of congestive heart failure cases to those with 'Both' heart failure types
if count_heart_failure_Both > 0:
    ratio = count_congestive_heart_failure / count_heart_failure_Both
else:
    ratio = None  # Avoid division by zero if no 'Both' heart failure type exists

print("Ratio of congestive heart failure to 'Both' heart failure types:", ratio)
```

```python
# Alternate solution
# Use filter() to create a filtered list of records where congestive heart failure is 1
congestive_heart_failure_cases = list(filter(lambda row: row['congestive_heart_failure'] == 1, df_CardiacComplication.to_dict('records')))

# Use filter() to create a filtered list of records where type of heart failure is 'Both'
heart_failure_Both_cases = list(filter(lambda row: row['type_of_heart_failure'] == 'Both', df_CardiacComplication.to_dict('records')))

# Count the number of cases for each
count_congestive_heart_failure = len(congestive_heart_failure_cases)
count_heart_failure_Both = len(heart_failure_Both_cases)

# Calculate the ratio
ratio = count_congestive_heart_failure / count_heart_failure_Both if count_heart_failure_Both > 0 else None

print("Ratio of congestive heart failure to 'Both' heart failure types:", ratio)
```

## 78 Give the count of patients who have experinced tachicardia or bradycardia.



```python
# Tachycardia (tak-ih-KAHR-dee-uh) is a fast heartbeat. The heart rate is greater than 100 beats a minute.
#Bradycardia (brad-e-KAHR-dee-uh) is a slow heartbeat. #The heart rate is less than 60 beats a minute.
# Your pulse rate, also known as your heart rate, is the number of times your heart beats per minute.
```

```python
# Define conditions for tachycardia and bradycardia
condition_tachycardia = df_labs['pulse'] > 100  # Tachycardia condition
condition_bradycardia = df_labs['pulse'] < 60  # Bradycardia condition

# Combine the conditions using bitwise OR ('|')
condition_tachycardia_bradycardia = condition_tachycardia | condition_bradycardia

# Filter the DataFrame to get patients with tachycardia or bradycardia
tachycardia_bradycardia_patients = df_labs[condition_tachycardia_bradycardia]

# Get the count of these patients
count_patients_with_both_conditions = tachycardia_bradycardia_patients['inpatient_number'].count()
print('count of patients who have experinced tachicardia or bradycardia :')
count_patients_with_both_conditions

```

## 79 Create a Funnel chart of BMI Category(Normal Weight, Underweight and Obese)



```python
!pip install plotly  
import plotly.express as px
# as per question but additional overweight
# Define a function to categorize BMI
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal Weight'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

# Apply the categorization function to create a new column 'BMI_category'
df_Demography['BMI_category'] = df_Demography['BMI'].apply(categorize_bmi)

# Get the count of each BMI category
bmi_counts = df_Demography['BMI_category'].value_counts().reset_index()
bmi_counts.columns = ['Category', 'Count']

# Create a funnel chart using Plotly Express
fig = px.funnel(
    bmi_counts,
    x='Count',
    y='Category',
    title='BMI Funnel Chart',
    labels={'Count': 'Number of Patients', 'Category': 'BMI Category'},
)

fig.show()
```

## 26 Create a column that randomly generates an age between the age group given for each patient. What was the average age of the patients?



```python
import random
import numpy as np
```

```python
df_demo_dupli = pd.read_excel(r"C:\Users\Home\Downloads\Data files\Python Hackethon may 2024\Cardiac_Outcomes.xlsx",
                                         sheet_name = 'Demography')
```

```python
import random
import numpy as np
# Function to generate a random age within a given age range
def generate_random_age(age_range):
    # Split the ageCat string to get the minimum and maximum ages
    age_min, age_max = map(int, age_range.split('-'))
    # Return a random integer between the min and max ages, inclusive
    return np.random.randint(age_min, age_max + 1)
#  use 'assign' to add a new column
df_demo_dupli_age = df_demo_dupli.assign(age=df_demo_dupli['ageCat'].apply(generate_random_age))
df_demo_dupli_age
average_age = round(df_demo_dupli_age['age'].mean())
print('Average age: ',average_age)

```
